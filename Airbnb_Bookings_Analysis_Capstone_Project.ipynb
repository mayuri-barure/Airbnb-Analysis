```python
avg_price_nbd= df.groupby(['neighbourhood_group','room_type'],dropna=True)['price'].mean().reset_index()
avg_price_nbd
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(df.longitude,df.latitude,hue=df.neighbourhood_group)
plt.ioff()
```

# Price Graph

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df[df['price'] < 300], x='longitude', y='latitude', hue='price', size="price")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='neighbourhood_group',y='price', hue='room_type', data=avg_price_nbd)
plt.show()
```

```python
plt.figure(figsize=(9,7))
mean_price_df = df.groupby('room_type', as_index=False)[['price']].mean()
sns.barplot(data=mean_price_df, x='room_type', y='price', palette='GnBu_d')
```

```python
n = df.neighbourhood_group.unique()
nbd_grp = list(n)
```

```python
for i in nbd_grp:
  nbd_price= df[df['neighbourhood_group']==i].groupby(['neighbourhood'])['price'].mean().reset_index(drop=False)
  nbd_price = pd.DataFrame(nbd_price)
  plt.figure(figsize=(18,5))
  plt.xticks(rotation=90)
  plt.title(i)
  plt.bar(nbd_price['neighbourhood'],nbd_price['price'])
```

# Availability Graph

```python
plt.figure(figsize=(10,10))
ax = sns.boxplot(data=df, x='neighbourhood_group',y='availability_365',palette='plasma')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(24, 8))
ax = axes.flatten()
sns.lineplot(data=df, x='neighbourhood_group', y='availability_365', hue='room_type', ax=ax[0])
ax[0].set_title('Room Availability throughout Neighbourhood/Room Type')
sns.scatterplot(data=df[df['price'] < 500], x="availability_365", y='price', hue='room_type', alpha=.9, palette="muted", ax=ax[1])
ax[1].set_title('Price vs Availability (Range $ 0 - 500)')

sns.countplot(data=df[df['availability_365']  == 365],x='neighbourhood_group', hue='room_type', palette='GnBu_d', ax=ax[2])
ax[2].set_title('Property Available 365 days')
sns.despine(fig)
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(df.longitude,df.latitude,hue=df.availability_365)
plt.ioff()
```

```python
for i in nbd_grp:
  nbd_available_rooms= df[df['neighbourhood_group']==i].groupby(['neighbourhood'])['availability_365'].mean().reset_index(drop=False)
  nbd_available_rooms = pd.DataFrame(nbd_available_rooms)
  plt.figure(figsize=(18,5))
  plt.xticks(rotation=90)
  plt.title(i)
  plt.bar(nbd_available_rooms['neighbourhood'],nbd_available_rooms['availability_365'])
```

# Reveiws Graph

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='reviews_per_month', size="reviews_per_month", sizes=(20, 150), hue_norm=(0, 5))
```

```python
for i in nbd_grp:
  nbd_number_of_reviews = df[df['neighbourhood_group']==i].groupby(['neighbourhood'])['number_of_reviews'].mean().reset_index(drop=False)
  nbd_number_of_reviews = pd.DataFrame(nbd_number_of_reviews)
  plt.figure(figsize=(18,5))
  plt.xticks(rotation=90)
  plt.title(i)
  plt.bar(nbd_number_of_reviews['neighbourhood'],nbd_number_of_reviews['number_of_reviews'])
```

# Famous Host

```python
plt1 = sns.barplot(x='Host_id',y='Count',data=famous_host)
plt1.set_xticklabels(plt1.get_xticklabels(), rotation=90)
```

```python
top_host_id = famous_host['Host_id'][famous_host['Count']==famous_host['Count'].max()]
top_host_name = df[['host_name','neighbourhood_group']][df['host_id']==top_host_id[0]].head(1)
top_host_name
```
